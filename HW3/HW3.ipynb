```python
group = happy.groupby(by="Region")
```

```python
group.mean()
```

```python
print("the most generous region is ")
happyregion = group.mean()
```

```python
ans = happyregion.loc[happyregion["Generosity"].idxmax()]
```

```python
print("the most generous region is " + ans.name)
```

```python
from scipy.stats import ks_2samp
```

```python
ks_2samp(happy["Generosity"], happy["Generosity"])
```

Comparing one sample to the same sample produces a p value of 1. a P value of 1 makes sense in this scenario as the two samples are identical and from the same group

```python
happy.corr()
```

```python
ks_2samp(happy["Generosity"], happy["year"])
```

Comparing Generosity and year, there is a p-value of .05. We can say that generosity and year are strongly correlated
